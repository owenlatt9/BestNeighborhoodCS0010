### Neighborhood Parks
Dataset Used: https://data.wprdc.org/dataset/parks/resource/fa329e3d-89ff-4708-8ddf-81bfedcad11d
This notebook is used to help determine the best Pittsburgh neighborhood for kids. More specifically, it determines the neighborhoods with the most parks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas
%matplotlib inline

```

This calculates the number of parks in each neighborhood.

```python
parks = pd.read_csv("parks.csv")
ppn = parks.groupby("neighborhood").size()
ppn = ppn.reset_index(name="Park Count")
print(ppn)
```

```python
ppn_sorted = ppn.sort_values(by='Park Count', ascending=False)
ppn_sorted.head(15)
```

```python
ppn_15= ppn_sorted.head(15)
ppn_15.plot(kind='bar', x ='neighborhood', y='Park Count', legend=False)
plt.title("15 Neighborhoods With The Most Parks")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Parks")
plt.show()
```
